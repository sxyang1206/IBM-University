```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import pandas as pd
import time
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 20)

# the difference is you need to put the review url e.g.https://studenthut.com/universities/university-leeds/reviews
review_url = "https://studenthut.com/universities/university-glasgow/reviews"

driver.get(review_url)

df_reviews = pd.DataFrame(columns=['University', 'Review'])
```

```python
#for just on page of review, run this~~
reviews_selector = ".sh-c-cards__item" #Selector

reviews = WebDriverWait(driver, 100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, reviews_selector)))

review_data = []
for review in reviews:
    review_text = review.text
    if review_text.strip():
        review_data.append({
            'University': 'University of Glasgow',
            'Review': review_text.strip()
        })

df_reviews = pd.DataFrame(review_data)



df_reviews.to_csv('/content/drive/MyDrive/Colab Notebooks/University of Glasgow_ratings.csv', index=False)

print(df_reviews)
```

```python
#for scrapping all review, run this

page_num = driver.find_elements(By.XPATH,'//*[@id="reviews"]/div/div[2]/div/a')[-1].text


reviews_selector = ".sh-c-cards__item" #Selector
review_data = []

for i in range(1, int(page_num) + 1):
    new_url = review_url + '?page={}'.format(i)
    driver.get(new_url)

    reviews = WebDriverWait(driver, 100).until(EC.presence_of_all_elements_located((By.CSS_SELECTOR, reviews_selector)))


    for review in reviews:
        review_text = review.text
        if review_text.strip():
            review_data.append({
                'University': 'University of Glasgow',
                'Review': review_text.strip()
            })

df_reviews = pd.DataFrame(review_data)


df_reviews.to_csv('/content/drive/MyDrive/Colab Notebooks/University of Glasgow_ratings.csv', index=False)

print(df_reviews)
```

```python
chrome_options = webdriver.ChromeOptions()
chrome_options.add_argument('--headless')
chrome_options.add_argument('--no-sandbox')
chrome_options.add_argument('--disable-dev-shm-usage')
driver = webdriver.Chrome(options=chrome_options)
wait = WebDriverWait(driver, 20)
name_list = []
for i in range(1,7,1):
  new_url = 'https://studenthut.com/universities' + '?page={}'.format(i)
  driver.get(new_url)
  name = driver.find_elements(By.XPATH,'/html/body/div[3]/div[3]/div/div[2]/article/a')
  name = [x.get_attribute('href')[23:] for x in name]
  name_list.extend(name)
```
